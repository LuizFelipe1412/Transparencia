```python
import pandas as pd

diretorio = '/home/luiz/Documentos/Transparência/Transparencia/Receitas/Planilhas'

dataframes = []

planilhas = ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL', 'MAIO', 'JUNHO']

for arquivo in planilhas:
    caminho_arquivo = f"{diretorio}/{arquivo}-2023 - RECEITA POR UNIDADE GESTORA.xlsx"
    
    if arquivo == 'MAIO':
        skiprows=5
    elif arquivo == 'FEVEREIRO':
        skiprows=8
    else:
        skiprows=7
        
    df = pd.read_excel(caminho_arquivo, skiprows=skiprows, skipfooter=2)
    df['Mes'] = arquivo.capitalize()
    dataframes.append(df)

df_receitas = pd.concat(dataframes)
    
df_receitas.rename(columns={'Unnamed: 0': 'Unidade_Gestora'}, inplace=True)
```
